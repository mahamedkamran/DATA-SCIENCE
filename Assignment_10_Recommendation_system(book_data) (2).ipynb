```python
import pandas as pd
import numpy as np
```

```python
book_data=pd.read_csv('book.csv')
book_data.head()
```

```python
book_data=book_data.rename({'User.ID':'userid','Book.Title':'book_title','Book.Rating':'book_rating'},axis=1)
book_data.head()
```

```python
len(book_data.book_title.unique())
```

```python
len(book_data.userid.unique())
```

```python
user_book_df = book_data.pivot_table(index='userid',
                                 columns='book_title',
                                 values='book_rating').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.fillna(0, inplace=True)
user_book_df.head()
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.index = book_data.userid.unique()
user_sim_df.columns = book_data.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
